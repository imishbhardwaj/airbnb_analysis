```python
import re

def remove_minutes(time_str):
    if isinstance(time_str, str):  # Check if time_str is a string
        # Pattern to match times with minutes (e.g., '12 00 pm', '1 00 pm')
        time_pattern = r'(\d{1,2}) (\d{2}) (am|pm)'

        # Replace the pattern with the format 'hour am/pm'
        cleaned_time = re.sub(time_pattern, r'\1 \3', time_str)

        return cleaned_time
    else:
        return time_str  # Return the original value if not a string

# Apply the function to the entire column
df['checkin'] = df['checkin'].apply(remove_minutes)
df['checkout'] = df['checkout'].apply(remove_minutes)

df.head()
```

**Checking for null values**

```python
df.isnull().sum()
```

**Filling missing values with test statistics**

```python

df['checkin'].replace(to_replace=np.nan,value=df['checkin'].mode()[0],inplace=True)
df['checkout'].replace(to_replace=np.nan,value=df['checkout'].mode()[0],inplace=True)
df.head()
```
